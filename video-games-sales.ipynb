### Data Cleaning

```python
missing_value = pd.DataFrame({'Missing_Values': df.isnull().sum(), "Percentage":df.apply(lambda x: f'{((x.isnull().sum()/df.shape[0])*100).round(2)}%')})
missing_value
```

```python
df.duplicated().sum()
```

```python
df_clean = df.copy()
df_clean.dropna(inplace=True)
```

```python
df_clean.isnull().sum()
```

```python
df_clean['Year'] = df_clean['Year'].astype(np.int64)
df_clean.info()
```

```python
df_clean
```

```python
df_clean.to_csv('video-games-sales.csv')
```
